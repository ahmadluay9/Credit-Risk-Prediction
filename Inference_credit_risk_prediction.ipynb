# Model Inference

## Import Libraries

```python
# Import Libraries
import pandas as pd
import numpy as np
import pickle
import json

# set_options
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_colwidth", 1000)
```

## Model Loading

```python
# Loading Model Files

with open('best_model.pkl', 'rb') as file_1:
  best_model = pickle.load(file_1)

with open('sel_features.txt','r') as file_2:
  sel_features = json.load(file_2)

with open('low_risk.txt','r') as file_3:
  low_risk = json.load(file_3)
```

## Predicting Data Inference

```python
# Load Data Inference
data_inf = pd.read_csv('data_inference.csv')
data_inf
```

```python
# Removing unnecessary features
data_inf_final = data_inf[sel_features]
round(data_inf_final,2)
```

```python
# Predict using Best Model
y_pred_inf = best_model.predict(data_inf_final) # pipeline_xgb from load files
y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['credit_risk_prediction'],index=data_inf_final.index)
round(y_pred_inf_df.head(25),2)
```

```python
# Actual
data_inf['actual_loan_status'] = data_inf['loan_status']
data_inf['actual_credit_risk'] = np.where(data_inf['actual_loan_status'].isin(low_risk), 'Low-Risk', 'High-Risk')
actual_inf = data_inf[['actual_loan_status','actual_credit_risk']]
actual_inf
```

```python
# Comparing the Actual loan_status and credit_risk_prediction
model_inf = pd.concat([actual_inf, y_pred_inf_df['credit_risk_prediction']], axis=1).sort_index()
model_inf['credit_risk_prediction'] = model_inf['credit_risk_prediction'].replace({1: 'Low-Risk', 0: 'High-Risk'})
model_inf
```

The insight from the given data is that the `actual_credit_risk` and `credit_risk_prediction` columns are mostly consistent, indicating that the model's credit risk predictions align with the actual loan status.

- For instances with `actual_loan_status` as 'Fully Paid', the `actual_credit_risk` and `credit_risk_prediction` are both 'Low-Risk', showing a correct prediction by the model.
- For instances with `actual_loan_status` as 'Charged Off', the `actual_credit_risk` and `credit_risk_prediction` are both 'High-Risk', again demonstrating accurate predictions by the model.
- There are some instances with `actual_loan_status` as 'Current', where the model has predicted 'Low-Risk', which suggests the model might be predicting a lower risk for ongoing loans.

Overall, the credit risk predictions made by the model (in the `credit_risk_prediction` column) seem to align well with the actual credit risk status (in the `actual_credit_risk` column) for the majority of the data points, indicating a reasonable performance of the classification model.
